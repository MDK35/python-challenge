```python
#Import Dependencies to read file 
import os 

import csv


#Load and output data from csv 

fileload = os.path.join(".", "Resources", "budget_data.csv")

fileoutput = os.path.join(".", "budget_analysis.txt")


total_mon = 1

total_net = 0

#Create an empty variable to store the values in
net_change_values = [ ]
change_in_month = [ ]

greatest = [" ", 0 ]
least = [" ",999999999999999999]

#Test of Loading
# print(fileload)
# print(fileoutput)

# Reading the CSV and converting it into a list
with open(fileload) as findata:
    scanner = csv.reader(findata)
    
    #Show the header row
    columns = next(scanner)
    
    print(f"columns: {columns}")
    first_row = next(scanner)
    
    #convert the first row to an integer 
    total_net += int(first_row[1])   
    prev_net = int(first_row[1])
    

    #create a loop to iterate throw rows
    for row in scanner:
        total_mon = total_mon + 1
        total_net += int(row[1])
        
        
    
        
        # Track the net change
        # subtract the previous values in the row to get the net change
        net_change = int(row[1]) - prev_net
        prev_net = int(row[1])
        # add the values into the empty list 
        net_change_values.append(net_change)
              
        
        # Calculate the greatest increase 
        if(net_change > greatest[1]):
            greatest[0] = row[0]
            greatest[1] = net_change
        
        # Calculate the greatest decrease
        if(net_change < least [1]):
            least[0] = row[0]
            least[1] = net_change




avg_net_monthly = sum(net_change_values)/ len(net_change_values)
        
output = (
    f"Total Months: {total_mon}\n"
    f"Total: ${total_net}\n"
    f"Average Change: ${avg_net_monthly:.2f}\n"
    f"Greatest Increase: {greatest}\n"
    f"Greatest Decrease: {least}\n"
)
print("$Financial Analysis$")    
print("-" * 25)
print(output)




with open(fileoutput, "w") as txt_file:
    txt_file.write(output)
```
